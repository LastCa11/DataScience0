```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame as "df"

# Filter the DataFrame to exclude rows where 'days_b_screening_arrest' is null
df_filtered = df.dropna(subset=['days_b_screening_arrest'])

# Create the scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['days_b_screening_arrest'], df_filtered['age'], alpha=0.5)
plt.title('Jail Time After First Charge vs. Age')
plt.xlabel('Days in Jail After First Charge')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame as "df"

# Filter the data to remove rows with negative values for 'days_b_screening_arrest' and 'age'
filtered_df = df[(df['days_b_screening_arrest'] >= 0) & (df['age'] >= 0)]

# Create the scatterplot
sns.set_theme()
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.scatterplot(data=filtered_df, x='days_b_screening_arrest', y='age')
plt.xlabel('Days in Jail')
plt.ylabel('Age')
plt.title('Scatterplot: Jail Time After First Charge vs. Age')

# Show the plot
plt.show()
```

```python
sns.set_theme()
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size
sns.countplot(data=df, x='decile_score')

# Add labels and title
plt.xlabel('COMPAS Score')
plt.ylabel('Count')
plt.title('Distribution of COMPAS Scores')

# Show the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame as "df"

# Create a bar/count plot to compare decile scores by sex
sns.set_theme()
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.countplot(data=df, x="decile_score", hue="sex")
plt.title("Comparison of Decile Scores by Sex")
plt.xlabel("Decile Score")
plt.ylabel("Count")
plt.legend(title="Sex", loc="upper right")  # Add a legend

plt.show()
```

```python
    import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame as "df"

# Filter out rows where both 'decile_score' and 'c_days_from_compas' are not null
filtered_df = df.dropna(subset=['decile_score', 'c_days_from_compas'])

# Create a scatter plot
sns.scatterplot(data=filtered_df, x='decile_score', y='c_days_from_compas')

# Set plot labels and title
plt.xlabel('Decile Score')
plt.ylabel('Days from Compas Screening to Jail')
plt.title('Relationship between Decile Scores and Time in Jail')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame as "df"

# Create a countplot to visualize the recidivism rate for each decile score
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="decile_score", hue="two_year_recid", palette="Set1")
plt.title("Recidivism Rate by Decile Score")
plt.xlabel("Decile Score")
plt.ylabel("Count")
plt.legend(title="Two-Year Recidivism", labels=["No", "Yes"])
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
