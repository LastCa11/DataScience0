```python
import zipfile
import os
import pandas as pd

# Define the name of the ZIP archive
zip_file_name = 'archive.zip'

# Create a temporary directory to extract the ZIP contents
extracted_dir = 'temp_extracted_data'
os.makedirs(extracted_dir, exist_ok=True)

# Extract the contents of the ZIP archive
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)

# List the files in the extracted directory
extracted_files = os.listdir(extracted_dir)

# Assuming you want to display the first CSV file found in the extracted directory
for file_name in extracted_files:
    if file_name.lower().endswith('.csv'):
        csv_file_path = os.path.join(extracted_dir, file_name)
        df = pd.read_csv(csv_file_path)
        print(f"First 5 rows of {file_name}:")
        print(df.head())
        break  # Exit the loop after the first CSV file is processed

# Clean up: Remove the temporary extracted directory
os.rmdir(extracted_dir)
```
